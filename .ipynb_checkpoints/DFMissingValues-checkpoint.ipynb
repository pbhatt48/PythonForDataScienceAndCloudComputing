```python
df = pd.read_csv('coffee_nmiss.csv')
df.head()
```

```python
missing = df.isnull().sum()
missing
```

```python
print(missing[missing > 0])
```

```python
list(missing[missing>0].values)
```

```python
list(missing[missing>0].index)
```

```python
df_fix = df.copy()
var_list = list(missing[missing>0].index)
for varnmae in var_list:
    misdummy = varnmae + '_misdummy'
    df_fix[misdummy] = pd.isnull(df_fix[str(varnmae)]) +0
df_fix.head()
```

```python
student_list = pd.read_csv('coffee_nmiss.csv')
student_list.head()
```

```python
from sklearn.preprocessing import Imputer
varlistnum = ['Person_ID','Age', 'Income', 'cup_tea']
dfnum = student_list[varlistnum]
dfnum.head(10)
```

```python
dfnum.isnull().sum()
```

```python
imp = Imputer(missing_values='NaN', strategy='mean', axis=0)
df_fix_oth = pd.DataFrame(imp.fit_transform(dfnum))
df_fix_oth.columns = varlistnum
df_fix_oth.head(10)
```

```python
# i want to know if there are any duplicates id
student_list.shape[0] - student_list['Person_ID'].nunique()
print(student_list.shape[0])
print(student_list['Person_ID'])
print(student_list['Person_ID'].nunique())

```

```python
#So basically i am substracting the unique numbers of Person ID from student list from the number of row
no_of_repeteated_personID = student_list.shape[0] - student_list['Person_ID'].nunique()
no_of_repeteated_personID
```

```python
#get duplicates for each id
dup = student_list.groupby('Person_ID').size()
dup.head()
```

```python
#get dup greater than 1
dup = dup[dup >1]
dup
```

```python
print('duplicated id', dup)
```

```python
#another way for removing the duplicates by keeping first occurance
student_list_unique_id = student_list.drop_duplicates('Person_ID', keep='first')
student_list_unique_id.head()
```
