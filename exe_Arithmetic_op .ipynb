```python
# This program is to show arithmetic operation on pandas DF
# header = 0 is default meaning use the first line as variable names in DF

import numpy as np
import pandas as pd

path=r'people_miss.csv'
df=pd.read_csv(path, usecols=['spend','income'], na_values='NA', header =0 )
df.head()
```

```python
df2 = pd.read_csv('people_miss.csv', usecols=['spend', 'income'], na_values='NaN', header=0)
df2
```

```python
# use a constant to multiply DF
(df*1.5).head()
```

```python
df3 = (df2 * 1.5).head()
df3
```

```python
# use a series to plus DF
addsome = pd.Series([300, 5000], index=['spend', 'income'])
(df + addsome).head()
```

```python
df3 +pd.Series([100, 200], index=['spend','income'])
```

```python
# use a DF to multiply add another DF
recordssize = len(df)
a1 = [1.35]*recordssize
a2 = [1.5]*recordssize
delta = pd.DataFrame({'spend': a1, 'income': a2})
print(a1, a2)
delta.head()
```

```python
x, y = [1] * len(df3), [2] *len(df3)
print(df3)
x, y
print(df3 * pd.Series([x , y], index=['spend', 'income']))
mult = pd.DataFrame({'spend':x, 'income':y})
print(df3 * mult)
```
