```python
S = pd.read_csv('coffee_info.csv')
mygroup = ['Gender', 'Employment']
mycolumn = ['Income', 'Cups_Per_Week','Age']
statistics = ['mean', 'std', 'max', 'min']
G = S.groupby(mygroup)
result = G[mycolumn].agg(statistics)
result
```

```python
np.round(result, decimals=1)
```
