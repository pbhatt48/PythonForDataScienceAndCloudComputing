```python
import numpy as np
import pandas as pd
```

```python
S = pd.read_csv('stock.csv', na_values='.')
S
```

```python
S['load_date'] = '2013-12-01'
S.head()
```

```python
S['diff_per'] = 100* (S.pc_1.fillna(0.0).astype(float) - S.pc_2.fillna(0.0).astype(float)) / (S.pc_2 + 0.0001)
```

```python
S[['exec', 'symbol', 'pc_1', 'pc_2', 'diff_per']].head()
```

```python
# append a new column 'top' and assign values 'Y' to rownames  1 , 2, 3 and 4
S.loc[1:6, 'top'] ="Y"
S[['exec', 'symbol','pc_1', 'pc_2', 'diff_per', 'top']].head(10)
```

```python
# append a new column ODDS and assign values to row (index) names  1 , 3 and 5
S.loc[[1,3,5], 'odds'] =1
S[['exec', 'symbol','pc_1', 'pc_2', 'diff_per', 'top', 'odds']].head(8)
```

```python
S.loc[S.pc_1<20, 'Ratio'] = (S.pc_1 - S.pc_2)/ S.pc_1
S.loc[(S.pc_1 <50) & (S.pc_1 >15), 'diff'] = S.pc_1 - S.pc_2
S[['exec', 'symbol','pc_1', 'pc_2', 'diff', 'Ratio']].head(8)
```

```python
S['avg3a'] = (S.pc_1 + S.pc_2 + S.pc_3) /3
S['avg3b'] = (S.pc_4 + S.pc_5 + S.pc_6) /3
S1 = S[((S.avg3a / S.avg3b) > 1.03)]
S2 = S[((S.avg3a /S.avg3b < 0.97))]
S1.head(8)
```
