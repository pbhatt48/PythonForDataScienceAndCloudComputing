```python
# show how stack() works
symbols=['stock1', 'stock2', 'stock3']
time=['day1','day2']
pricedata=[[11.3, 9.3, 20.5], [11.8, 9.8, 19.3]]
df = pd.DataFrame(pricedata, index=time, columns=symbols)
df
```

```python
df.stack()
```

```python
# we show how to exchange row and column using stack and unstack methods
symbols=['stock1', 'stock2', 'stock3']
time=['day1','day2']
pricedata=[[11.3, 9.3, 20.5], [11.8, 9.8, 19.3]]
stockprice =pd.DataFrame(pricedata, columns=symbols, index=time)
print (stockprice)

stockpricest=stockprice.stack()
stockpricest
```

```python
 # the level=0 represents the first index: day1, day2...
X = stockpricest.unstack(level=0)
X
```

```python
 # # the level=1 represents the second index: stock1, stock2.
Y = stockpricest.unstack(level=1)
Y
```

```python
# we show how to exchange row and column using melt and pivot methods
symbols=['time','stock1', 'stock2', 'stock3']
pricedata=[['day1',11.3, 9.3, 20.5], ['day2',11.8, 9.8, 19.3]]
stockprice =pd.DataFrame(pricedata, columns=symbols)
print (stockprice)
```

```python
X = pd.melt(stockprice, id_vars=['time'], value_vars=['stock1','stock2','stock3'])
X
```

```python
Y = X.pivot(index='variable',columns='time',values='value').reset_index()
Y
```

```python
# you can also use unstack(), but should be indexed first
X=pd.melt(stockprice ,id_vars=['time'],value_vars=['stock1', 'stock2', 'stock3'])
Z=X.set_index(['time', 'variable'])  
Z.unstack(level=1)
Z.unstack(level=0).reset_index()
print (Z)
```

```python
retail = pd.read_csv('retail_transactions.csv', na_values='NA')
retail.head()
```

```python
retail.columns.values
```

```python
df1 = retail[['Customer_ID', 'Transaction_ID', 'Store_Department']]
Z = df1.groupby(['Customer_ID', 'Store_Department']).count()
Z
```

```python
Z = Z.add_suffix('_Count').reset_index()
Z
```

```python
M=Z.pivot(index='Customer_ID',columns='Store_Department',values='Transaction_ID_Count').head()
M=M.reset_index()
M.head()
```

```python
deplist=list(set(retail.Store_Department))
M.set_index(['Customer_ID'])
N=pd.melt(M, id_vars=['Customer_ID'], value_vars=deplist)
N=N[N.value.isnull()==False]
N.head(10)
```
