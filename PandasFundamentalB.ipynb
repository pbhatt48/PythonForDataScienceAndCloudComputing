```python
#read large files in chunks
reader = pd.read_csv('Review_b.csv', iterator=True)
loop = True
chunks = []
chunk_size = 100000
while loop:
    try:
        chunk = reader.get_chunk(chunk_size)
        chunks.append(chunk)
    except StopIteration:
        loop = False
        print("Iteration is Stopped!")
```

```python
len(chunks)
```

```python
#merger all the chunks
df = pd.concat(chunks, ignore_index=True)
print(df.columns)
df.head()
```
