```python
import numpy as np
import pandas as pd
```

```python
#get the sheet name
pdexcel = pd.ExcelFile('people_miss.xlsx')
print(pdexcel.sheet_names)
```

```python
#open the sheet of x name
dfexcel = pdexcel.parse('people_nmiss')
dfexcel.head()
```

```python
#add a column 
dfexcel['utilization'] = dfexcel.spend / dfexcel.upperlimt
dfexcel.income = dfexcel.income * 0.69
dfexcel.head()
```

```python
#add a column with np.where dataframe
dfexcel['young'] = np.where((dfexcel.age <30),True, False)
dfexcel.head()
```

```python
#drop the added column
dfexcel.drop(columns=['young'], inplace=True)
dfexcel.head()
```

```python
#insert the same column at certain index
dfexcel.insert(3,'young',(dfexcel.age < 30))
dfexcel.head()
```

```python
#get the map series for the age group
newdf = dfexcel.copy()
age_series = newdf.age
def getagestatus(age):
    if age < 35:
        return '1'
    return '0'

agetarget = age_series.map(getagestatus)
agetarget.head()
```

```python
#use apply function for the DF
newdf = dfexcel.copy()
def change_income(D):
    if D.education == 'U':
        x = D.income + 1000
    else:
        x = D.income + 500
    return x

newdf['new_income'] = newdf.apply(change_income, axis=1)
v = ['education', 'income', 'new_income']
newdf[v].head()
```

```python
dfexcel = pdexcel.parse('people_nmiss', na_values='NaN')
dfexcel.head(8)
```

```python
#practice the same above

newdf = dfexcel.copy()
def class_separation(D):
    if D.income < 15000:
        x = 'low'
    elif D.income >= 15000 and D.income <40000:
        x = 'middle'
    elif D.income == 'NaN':
        x = 'None'
    else:
        x = 'high'
    return x

newdf['class'] = newdf.apply(class_separation, axis=1)
newdf.head(10)
```
