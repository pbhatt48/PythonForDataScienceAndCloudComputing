```python
##ranking function
def grouping(data, groupnum):
    grp = pd.qcut(data['income'], groupnum, labels=False)
    data['ranks'] = grp
    return data

consumers = pd.read_csv('consumers_ext.csv', index_col='customer_id')
consumers.head(n=10)
```

```python
df_ranked = grouping(consumers,5)
df_ranked.head()
```
