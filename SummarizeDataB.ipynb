```python
# we get some sample for data analysis
sample_df = df.sample(n=16000, random_state=2, replace=False)
print(sample_df.shape)
sample_df.head()
```

```python
#what is the avgerage mortgage, rent_median and mortgage / rent_median over all the records?
sample_df['mor_rent_ratio'] = sample_df.mortgage/sample_df.rent_median
varlist = ['mor_rent_ratio', 'mortgage', 'rent_median']
sample_df.head()
```

```python
s = sample_df[varlist].mean(axis=0)
```

```python
type(s)
```

```python
city_avg = pd.DataFrame(s, columns=['Avg'])
city_avg = city_avg.round(1)
city_avg
```

```python
# #what is the avgerage mortgage, rent_median and mortgage / rent_median over all the states?

varlist = ['state','mor_rent_ratio', 'mortgage', 'rent_median']
gr_state_avg = sample_df.groupby('state')[varlist].mean().reset_index()
gr_state_avg = gr_state_avg.round({'mor_rent_ratio' : 1, 'mortgage' : 1, 'rent_median' : 1})
```

```python
# I want to know which city has the highest AVG mortgage and AVG mortgage/rent_median
var_list = ['mor_rent_ratio', 'mortgage']
D = sample_df.groupby('city')[var_list].mean()
print(D.sort_values(by=['mor_rent_ratio'], ascending=False).head(1))
print(D.sort_values(by=['mortgage'], ascending=False).head(1))
```

```python
D.idxmax()
```

```python
# I want to know which city has the highest AVG  population and their rent pay
var_list = ['num_population', 'rent_median']
X = sample_df.groupby('city')[var_list].mean()
X.sort_values(by='num_population', ascending=True).head()
X.idxmax()
```
