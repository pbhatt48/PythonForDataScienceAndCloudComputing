```python
df = pd.read_csv('people_miss.csv', na_values='.')
df.head()
```

```python
df.shape
```

```python
#to see if the cell's value is missing
Y = df.isnull()
Y.head()
```

```python
Z = df.notnull()
Z.head()
```

```python
# to see missing# for each variable 
Y = df.isnull().sum()
Y
```

```python
# count for counting the non missing variable
df.count()
```

```python
#drop rows with any missing value
A = df.dropna()
print("Before == ", df.shape)
print("After == ", A.shape)
A.head()
```

```python
#drop columns with any missing value
A = df.dropna(axis=1)
print("Before == ", df.shape)
print("After == ", A.shape)
A.head()
```

```python
#drop rows if all cells have missing value
A = df.dropna(how='all', axis=0)
print("Before == ", df.shape)
print("After == ", A.shape)
A.head()
```

```python
#drop columns if all cells have missing value
A = df.dropna(how='all', axis=1)
print("Before == ", df.shape)
print("After == ", A.shape)
A.head()
```

```python
#drop columns with thresold of atleast two empty values if all cells have missing value
#thresh=N requires that a column has at least N non-NaNs to survive.
df = pd.read_csv('people_miss.csv', na_values='NA')
A = df.dropna(thresh=9, axis=0)
print("Before == ", df.shape)
print("After == ", A.shape)
A.head(10)
```

```python
A = df.dropna(subset=['upperlimit','income'],how='any', axis=0)
print("Before == ", df.shape)
print("After == ", A.shape)
A.head(10)
```
