```python
df = pd.read_csv('people_miss.csv')
df
```

```python
df_target = pd.read_csv('people_target.csv', na_values='NA')
df_target
```

```python
alldf = pd.merge(df, df_target, on=['id'], how='outer')
alldf
```

```python
pd.isnull(alldf.target)
```

```python
#umnbr of true values
pd.isnull(alldf.target).sum()
```

```python
# of false values = non na target
(pd.isnull(alldf.target) == False).sum()
```

```python
# of non na target
1 - (alldf.target.isnull()).sum()
```

```python
pd.isnull(alldf['name']).sum()
```

```python
# how many NA target?
print ('# of na target', pd.isnull(alldf.target).sum())
# how many non NA target?
print ('# of non na target', (pd.isnull(alldf.target)==False).sum())
print ('# of non na target', (1-alldf.target.isnull()).sum())
# how many NA names?
print ('# of na name', pd.isnull(alldf['name']).sum())
# how much target rate for non NA targets
print ('target rate', alldf[alldf.target.isnull()==False]['target'].mean())
# how mucn NA target rate?
print ('NA target rate', pd.isnull(alldf['target']).mean())
# descriptive statistics for spend, income
print ('income max', alldf.income.max())
print ('income mean', alldf.income.mean())
print ('spend max', alldf.spend.max())
print ('spend mean', alldf.spend.mean())
```

```python
# corrlation between age, income, spend, sp_ratio with target
nonatarget=alldf[alldf.target.isnull()==False]
nonatarget=nonatarget[nonatarget.age.isnull()==False]
nonatarget=nonatarget[nonatarget.income.isnull()==False]
nonatarget=nonatarget[nonatarget.spend.isnull()==False]
nonatarget.head()
```
