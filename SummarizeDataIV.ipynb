```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('demograhics.csv', na_values="NA")
```

```python
sample_df = df.sample(n=16000, replace=False, random_state=2)
sample_df.head()
```

```python
v =['city', 'state']
X = sample_df[v].groupby(['state','city']).count().reset_index()
X.head()
```
